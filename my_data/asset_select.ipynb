```python
import requests
import pandas as pd
from datetime import datetime
import io
import zipfile

# 目标回测起始时间
TARGET_DATE = "2024-01-15" # 选一月中旬的一天作为“探针”，避开元旦假期可能的异常
TARGET_TIMESTAMP = datetime(2024, 1, 1).timestamp() * 1000 # 2024-01-01 的毫秒时间戳

def get_tradable_vectors():
    print("1. 获取所有当前交易对...")
    url = "https://fapi.binance.com/fapi/v1/exchangeInfo"
    resp = requests.get(url).json()
    
    candidates = []
    for s in resp['symbols']:
        # 筛选条件1: 必须是 USDT 本位合约
        # 筛选条件2: 必须在交易中 (TRADING)
        # 筛选条件3: 上线时间必须早于 2024-01-01
        if s['contractType'] == 'PERPETUAL' and \
           s['quoteAsset'] == 'USDT' and \
           s['status'] == 'TRADING' and \
           s['onboardDate'] < TARGET_TIMESTAMP:
            candidates.append(s['symbol'])
            
    print(f"   符合时间条件的‘老币’共有: {len(candidates)} 个")
    return candidates

def probe_volume(symbol):
    """
    下载该币 2024-01-15 一天的数据，计算成交额
    """
    url = f"https://data.binance.vision/data/futures/um/daily/klines/{symbol}/1m/{symbol}-1m-{TARGET_DATE}.zip"
    try:
        r = requests.get(url, timeout=5)
        if r.status_code != 200:
            return 0
        
        with zipfile.ZipFile(io.BytesIO(r.content)) as z:
            with z.open(z.namelist()[0]) as f:
                # 币安K线第6列是 Quote Volume (成交额)
                df = pd.read_csv(f, header=None)
                daily_volume = df.iloc[:, 7].sum() # 第8列(索引7)通常是 Quote Asset Volume
                return daily_volume
    except Exception:
        return 0

def main():
    # 1. 拿到所有“老币”
    candidates = get_tradable_vectors()
    
    print(f"2. 开始对 {len(candidates)} 个老币进行‘流量探针’测试 (2024-01-15的数据)...")
    print("   这可能需要几分钟，请耐心等待...")
    
    volume_rank = []
    
    # 这里为了演示不加多线程，如果你想快可以像之前那样并发
    total = len(candidates)
    for i, symbol in enumerate(candidates):
        vol = probe_volume(symbol)
        if vol > 0:
            volume_rank.append((symbol, vol))
        print(f"\r   进度: {i+1}/{total} - {symbol}: ${vol:,.0f}", end="")
    
    print("\n3. 排序完成！")
    
    # 按成交额倒序排列
    volume_rank.sort(key=lambda x: x[1], reverse=True)
    
    # 取前 200
    top_200 = [x[0] for x in volume_rank[:200]]
    
    print("\n" + "="*50)
    print("以下是 2024年1月15日 成交额最大的 200 个币种列表：")
    print("="*50)
    print(str(top_200))
    print("="*50)
    print(f"\n建议复制上面的列表，替换 download_data.py 中的 SYMBOLS 变量。")

if __name__ == "__main__":
    main()
```
