```python
import os
import pandas as pd
import glob
from tqdm import tqdm  # У┐Џт║дТЮАт║Њ№╝їтдѓТъюТ▓АТюЅт«ЅУБЁтЈ»С╗ЦСй┐уће pip install tqdm

def convert_parquet_to_gzip(root_folder):
    # жЂЇтјєТЅђТюЅтГљТќЄС╗Хтц╣
    # Сй┐уће os.walk ТѕќУђЁ glob (У┐ЎжЄїуће os.walk С╗ЦтцёуљєС╗╗ТёЈТи▒т║дуџётГљуЏ«тйЋ)
    file_list = []
    for root, dirs, files in os.walk(root_folder):
        for file in files:
            if file.endswith(".parquet") and not file.endswith(".parquet.gz"):
                file_list.append(os.path.join(root, file))

    print(f"ТЅЙтѕ░ {len(file_list)} СИф Parquet ТќЄС╗Х№╝їт╝ђтДІУйгТЇб...")

    for old_path in tqdm(file_list):
        try:
            # 1. У»╗тЈќтјЪТќЄС╗Х
            df = pd.read_parquet(old_path)
            
            # 2. ТъёжђаТќ░ТќЄС╗ХтљЇ (file.parquet -> file.parquet.gz)
            # Т│еТёЈ№╝џТюЅС║ЏС╣аТЃ»Тў» file.parquet ТюгУ║Фт░▒Тў»тјІу╝Еуџё№╝їСйєтдѓТъюСйаТЃ│ТўјуА«тљју╝ђ
            new_path = old_path + ".gz" 
            
            # 3. тєЎтЁЦТќ░ТќЄС╗Х№╝їТїЄт«џтјІу╝ЕТа╝т╝ЈСИ║ gzip
            df.to_parquet(new_path, compression='gzip')
            
            # 4. (тЈ»жђЅ) тѕажЎцтјЪТќЄС╗Х - т╗║У««уА«У«цУйгТЇбТЌаУ»»тљјтєЇтЈќТХѕТ│ежЄі
            # os.remove(old_path)
            
        except Exception as e:
            print(f"тцёуљєТќЄС╗Х {old_path} ТЌХтЄ║жћЎ: {e}")

# Сй┐ућеуц║СЙІ
if __name__ == "__main__":
    target_dir = r"./data/futures/um/daily/klines"  # С┐«Тћ╣У┐ЎжЄї
    convert_parquet_to_gzip(target_dir)
```

```python
import os

# ================= жЁЇуй«тї║тЪЪ =================
# ТЋ░ТЇ«Та╣уЏ«тйЋ (У»иСй┐ућеу╗Ют»╣Уи»тЙё)
BASE_DIR = r"./data/futures/um/daily/klines" 

# сђљжЄЇУдЂсђЉTrue = жбёУДѕТеАт╝Ј (тЈфТЅЊтЇ░СИЇтѕа); False = уюЪтѕаТеАт╝Ј
DRY_RUN = False
# ===========================================

def safe_clean_parquet():
    print(f"­ЪЊѓ ТЅФТЈЈуЏ«тйЋ: {BASE_DIR}")
    print(f"­ЪЏА№ИЈ ТеАт╝Ј: {'­ЪЉђ жбёУДѕТеАт╝Ј (т«ЅтЁе)' if DRY_RUN else '­ЪћЦ ТЅДУАїТеАт╝Ј (ТЁјжЄЇ)'}")
    print("-" * 50)
    
    deleted_count = 0
    skipped_count = 0
    freed_space = 0
    
    for root, dirs, files in os.walk(BASE_DIR):
        for file in files:
            # 1. СИЦТа╝уГЏжђЅуЏ«ТаЄ№╝џт┐ЁжА╗С╗Ц .parquet у╗Њт░Й
            # Т│еТёЈ№╝џPython уџё endswith(".parquet") СИЇС╝џтї╣жЁЇ ".parquet.gz"
            if file.endswith(".parquet"):
                
                parquet_path = os.path.join(root, file)
                gz_path = parquet_path + ".gz" # ТјеТхІт»╣т║ћуџё gz ТќЄС╗ХУи»тЙё
                
                # 2. т«ЅтЁеТБђТЪЦ№╝џтЈфТюЅт»╣т║ћуџё .gz ТќЄС╗ХтГўтюе№╝їТЅЇтЁЂУ«ИтѕажЎцтјЪТќЄС╗Х
                if os.path.exists(gz_path):
                    
                    # (тЈ»жђЅ) жбЮтцќТБђТЪЦ№╝џуА«С┐Ю .gz ТќЄС╗ХСИЇТў»уЕ║уџё
                    if os.path.getsize(gz_path) == 0:
                        print(f"Рџа№ИЈ [Уи│У┐Є] тЈЉуј░т»╣т║ћуџё .gz ТќЄС╗ХтцДт░ЈСИ║0№╝їС┐ЮуЋЎтјЪТќЄС╗Х: {file}")
                        skipped_count += 1
                        continue

                    try:
                        size = os.path.getsize(parquet_path)
                        freed_space += size
                        
                        if DRY_RUN:
                            print(f"[тЙЁтѕажЎц] {file} (РюЁ ти▓уА«У«цт»╣т║ћ .gz тГўтюе)")
                        else:
                            os.remove(parquet_path)
                            print(f"[ти▓тѕажЎц] {file}")
                            
                        deleted_count += 1
                        
                    except Exception as e:
                        print(f"РЮї [жћЎУ»»] тѕажЎцтц▒У┤Ц {file}: {e}")
                
                else:
                    # тдѓТъюТ▓АТюЅт»╣т║ћуџё .gz ТќЄС╗Х№╝їу╗Ют»╣СИЇУЃйтѕа№╝Ђ
                    print(f"­ЪЏЉ [С┐ЮуЋЎ] ТюфТЅЙтѕ░т»╣т║ћуџё .gz ТќЄС╗Х№╝їС┐ЮуЋЎтјЪТќЄС╗Х: {file}")
                    skipped_count += 1

    # у╗ЪУ«А
    size_mb = freed_space / (1024 * 1024)
    print("-" * 50)
    if DRY_RUN:
        print(f"жбёУДѕу╗ЊТЮЪ№╝џ")
        print(f"РюЁ У«АтѕњтѕажЎц: {deleted_count} СИфТќЄС╗Х (жЃйТюЅт»╣т║ћуџё .gz)")
        print(f"­ЪЏЉ У«АтѕњС┐ЮуЋЎ: {skipped_count} СИфТќЄС╗Х (Т▓АТЅЙтѕ░т»╣т║ћуџё .gz)")
        print(f"­ЪЊд жбёУ«АжЄіТћЙ: {size_mb:.2f} MB")
        print("\n­ЪЉЅ уА«У«цТЌаУ»»тљј№╝їт░єС╗БуаЂСИГуџё DRY_RUN Тћ╣СИ║ False тєЇТгАУ┐љУАїсђѓ")
    else:
        print(f"ТИЁуљєт«їТѕљ№╝џ")
        print(f"­ЪЌЉ№ИЈ т«ъжЎЁтѕажЎц: {deleted_count} СИфТќЄС╗Х")
        print(f"­ЪЏЉ т«ъжЎЁС┐ЮуЋЎ: {skipped_count} СИфТќЄС╗Х")
        print(f"­ЪЊд жЄіТћЙуЕ║жЌ┤: {size_mb:.2f} MB")

if __name__ == "__main__":
    safe_clean_parquet()
```
